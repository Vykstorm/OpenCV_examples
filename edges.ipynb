This notebook shows how to do edge detection using different methods

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv.cvtColor(cv.imread('images/taj-mahal.jpg'), cv.COLOR_RGB2GRAY)
```

```python
plt.imshow(img, cmap='gray'), plt.xticks([]), plt.yticks([]);
```

```python
kernel_size = 9
sobelx = cv.Sobel(img, cv.CV_64F, 1, 0, kernel_size)
sobely = cv.Sobel(img, cv.CV_64F, 0, 1, kernel_size)
laplacian = cv.Laplacian(img, cv.CV_64F, kernel_size)

fig, ax = plt.subplots(1, 3, figsize=(15, 6))

plt.sca(ax[0])
plt.imshow(sobelx, cmap='gray')
plt.title('Sobel Vertical'), plt.xticks([]), plt.yticks([])

plt.sca(ax[1])
plt.imshow(sobely, cmap='gray')
plt.title('Sobel Horizontal'), plt.xticks([]), plt.yticks([]);

plt.sca(ax[2])
plt.imshow(laplacian, cmap='gray')
plt.title('Laplacian'), plt.xticks([]), plt.yticks([]);
```
