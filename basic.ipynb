This notebook shows how to load/show/save images using openCV

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Load an image

```python
img = cv.imread('images/cat.jpg')
```

```python
img.shape
```

```python
img.dtype
```

```python
plt.imshow(img[:, :, (2, 1, 0)])
plt.xticks([])
plt.yticks([]);
```

## Image ROI

Access a specific image region using numpy indexing

```python
region = img[30:110,70:150,:]
plt.imshow(region[:, :, (2, 1, 0)])
plt.xticks([])
plt.yticks([]);
```

## Save the image

```python
cv.imwrite('images/cat-eye.png', region)
```
