This notebook shows different tresholding techniques using OpenCV <br>
You can read this <a href='https://docs.opencv.org/trunk/d7/d4d/tutorial_py_thresholding.html'>tutorial</a> for more info

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
img = cv.imread('images/room.jpg')
```

```python
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([]);
```

Convert the image color space to gray

```python
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
```

```python
img.shape
```

```python
plt.imshow(img, cmap='gray')
plt.xticks([]), plt.yticks([]);
```

## Simple thresholding 

```python
fig, ax = plt.subplots(1, 3, figsize=(16, 6))
thresholds = [65, 120, 180]

for i in range(0, 3):
    plt.sca(ax[i])
    
    ret, img2 = cv.threshold(img, thresholds[i], 255, cv.THRESH_BINARY)

    plt.imshow(img2, cmap='gray')
    plt.xticks([]), plt.yticks([])
    plt.title('Threshold: {}'.format(thresholds[i]))
plt.suptitle('Simple thresholding')
plt.tight_layout()
plt.subplots_adjust(top=1.2)
```

## Adaptative thresholding

```python
fig, ax = plt.subplots(2, 3, figsize=(16, 8))
C = [5, 10, 20]
kernels = [13, 25, 37]

for i in range(0, 3):
    plt.sca(ax[0, i])
    
    img2 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 13, C[i])

    plt.imshow(img2, cmap='gray')
    plt.xticks([]), plt.yticks([])
    plt.title('C={}, Kernel size={}'.format(thresholds[i], 13))
    
for i in range(0, 3):
    plt.sca(ax[1, i])
    
    img2 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, kernels[i], 10)

    plt.imshow(img2, cmap='gray')
    plt.xticks([]), plt.yticks([])
    plt.title('C={}, Kernel size={}'.format(10, kernels[i]))
    
plt.suptitle('Adaptative thresholding (mean)')
plt.tight_layout()
plt.subplots_adjust(top=0.85)
```

```python
fig, ax = plt.subplots(2, 3, figsize=(16, 8))
C = [5, 10, 20]
kernels = [13, 25, 37]

for i in range(0, 3):
    plt.sca(ax[0, i])
    
    img2 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 13, C[i])

    plt.imshow(img2, cmap='gray')
    plt.xticks([]), plt.yticks([])
    plt.title('C={}, Kernel size={}'.format(thresholds[i], 13))
    
for i in range(0, 3):
    plt.sca(ax[1, i])
    
    img2 = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, kernels[i], 10)

    plt.imshow(img2, cmap='gray')
    plt.xticks([]), plt.yticks([])
    plt.title('C={}, Kernel size={}'.format(10, kernels[i]))
    
plt.suptitle('Adaptative thresholding (gaussian)')
plt.tight_layout()
plt.subplots_adjust(top=0.85)
```

## Otsu thresholding

```python
threshold, img2 = cv.threshold(img, 0, 255, cv.THRESH_BINARY + cv.THRESH_OTSU)
'Threshold is {}'.format(threshold)
```
