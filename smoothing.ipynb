This notebook shows how to smooth images using different filters

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv

%matplotlib inline
```

```python
img = cv.imread('images/taj-mahal.jpg')
```

```python
img.shape
```

```python
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB)), plt.xticks([]), plt.yticks([]);
```

## Average filter

```python
kernel = np.ones([7, 7]) / 7 ** 2
```

```python
img2 = cv.filter2D(img, ddepth=-1, kernel=kernel)
```

```python
img2.shape
```

```python
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB)), plt.xticks([]), plt.yticks([]);
```

```python
img2 = cv.blur(img, (7, 7))
```

```python
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB)), plt.xticks([]), plt.yticks([]);
```

## Gaussian Blurring

```python
img2 = cv.GaussianBlur(img,(7,7),0)
```

```python
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB)), plt.xticks([]), plt.yticks([]);
```

## Median Blurring

```python
img = cv.imread('images/butterfly.png')
img2 = cv.medianBlur(img, 5)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6));
plt.sca(ax[0])
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB)), plt.xticks([]), plt.yticks([]);
plt.sca(ax[1])
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB)), plt.xticks([]), plt.yticks([]);
```

## Bilateral filter

```python
img = cv.imread('images/taj-mahal.jpg')
```

```python
img2 = cv.bilateralFilter(img, 9, 75, 75)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6));
plt.sca(ax[0])
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB)), plt.xticks([]), plt.yticks([]);
plt.sca(ax[1])
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB)), plt.xticks([]), plt.yticks([]);
```
