In this script we will show how to apply affine transformations to images using OpenCV

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
from math import cos, sin, pi, radians, degrees

%matplotlib inline
```

```python
img = cv.imread('images/cat.jpg')
```

```python
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([]);
```

## Scaling

```python
img.shape
```

```python
img2 = cv.resize(img, None, fx=0.25, fy=0.25, interpolation=cv.INTER_LINEAR)
```

```python
img2.shape
```

```python
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([]);
```

```python
img2 = cv.resize(img, (128, 128), interpolation=cv.INTER_AREA)
```

```python
img2.shape
```

```python
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([]);
```

## Translation

```python
M = np.array([
    [1, 0, 20],
    [0, 1, 50]
], dtype=np.float32)
img2 = cv.warpAffine(img, M, (img.shape[1], img.shape[0]))
```

```python
img2.shape
```

```python
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([]);
```

## Rotation

```python
M = cv.getRotationMatrix2D((200, 200), 45, 1)
img2 = cv.warpAffine(img, M, (img.shape[1], img.shape[0]))
```

```python
plt.imshow(cv.cvtColor(img2, cv.COLOR_BGR2RGB))
plt.xticks([]), plt.yticks([]);
```
